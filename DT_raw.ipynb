```python
import pandas
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
```

```python
df_train = pandas.read_csv('./data/train.csv')
df_test  = pandas.read_csv('./data/test.csv')  
```

```python
df_train.head()
```

```python
df_train.Activity.unique()
```

```python
df_test['Activity'] = pandas.factorize(df_test['Activity'])[0] + 1
df_train['Activity'] = pandas.factorize(df_train['Activity'])[0] + 1
```

```python
df_train.head()
```

```python
X = df_train.drop(columns=df_train.columns.tolist()[-1])
y = df_train[df_train.columns.tolist()[-1]]

classifier = DecisionTreeClassifier(min_impurity_decrease=0.001)
classifier.fit(X, y)
```

```python
X_test = df_test.drop(columns=df_test.columns.tolist()[-1])
y_test = df_test[df_test.columns.tolist()[-1]]

y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# https://stackoverflow.com/questions/49428469/pruning-decision-trees
from sklearn.tree._tree import TREE_LEAF

def prune_index(inner_tree, index, threshold):
    if inner_tree.value[index].min() < threshold:
        # turn node into a leaf by "unlinking" its children
        inner_tree.children_left[index] = TREE_LEAF
        inner_tree.children_right[index] = TREE_LEAF
    # if there are children, visit them as well
    if inner_tree.children_left[index] != TREE_LEAF:
        prune_index(inner_tree, inner_tree.children_left[index], threshold)
        prune_index(inner_tree, inner_tree.children_right[index], threshold)
```

```python
print(sum(classifier.tree_.children_left < 0))
prune_index(classifier.tree_, 0, 1)
```

```python
y_pred = classifier.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
