```python
import pandas
```

```python
df = pandas.read_csv('./data/test.csv')
```

```python
participants = {}

for index, row in df.iterrows():
    subject = row['subject']
    activity = row['Activity']
    
    if subject not in participants:
        participants[subject] = set([activity])
    else:
        participants[subject].add(activity)
        
```
